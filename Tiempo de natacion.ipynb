```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import stats
import math
from scipy import signal
```

```python
import pandas as pd
df = pd.read_csv("rioTimes.csv")
katieTimes = df[df["name"]=="Katie LEDECKY"]["time"].values
katieTimesm = katieTimes[3:-2]
katieTimesme=signal.detrend(katieTimesm)
KatieIda=np.array([value for key,value in enumerate(katieTimesme) if key%2==0])
KatieVuelta=[value for key,value in enumerate(katieTimesme) if key%2==1]

print stats.kurtosistest(KatieIda)
#print stats.skewtest(KatieIda)
#print stats.normaltest(KatieIda)
print stats.shapiro(KatieIda)
print stats.kstest(KatieIda, "norm")
stats.ttest_rel(KatieIda[:-1], KatieVuelta)
```
